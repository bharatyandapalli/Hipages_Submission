### TASK 1: SQL

### Query 1: Names and the Number of Messages Sent by Each User
```sql
SELECT 
    U.Name AS UserName,
    COUNT(M.MessageID) AS MessagesSent
FROM 
    User U
JOIN 
    Messages M ON U.UserID = M.UserIDSender
GROUP BY 
    1
    
```


### Query 2: Total Number of Messages Sent Stratified by Weekday
```
SELECT 
    DAYNAME(M.DateSent) AS Weekday,
    COUNT(M.MessageID) AS TotalMessages
FROM 
    Messages M
GROUP BY 
    Weekday
ORDER BY 
    FIELD(Weekday, 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday');```



### Query 3: Most Recent Message from Each Thread That Has No Response Yet

```
WITH UnresolvedMessages AS (
    SELECT 
        M.ThreadID,
        M.MessageID,
        M.MessageContent,
        M.DateSent,
        ROW_NUMBER() OVER (PARTITION BY M.ThreadID ORDER BY M.DateSent DESC) AS rn
    FROM 
        Messages M
    LEFT JOIN 
        Messages M2 
        ON M.ThreadID = M2.ThreadID 
        AND M.UserIDRecipient = M2.UserIDSender 
        AND M2.DateSent > M.DateSent
    WHERE 
        M2.MessageID IS NULL
)
SELECT 
    T.ThreadID,
    T.Subject,
    UM.MessageContent,
    UM.DateSent
FROM 
    Threads T
JOIN 
    UnresolvedMessages UM 
    ON T.ThreadID = UM.ThreadID
WHERE 
    UM.rn = 1;

    ```


### Query 4: Conversation with the Most Messages (All User Data and Message Contents)

```
WITH MostMessagesThread AS (
    SELECT 
        ThreadID
    FROM 
        Messages
    GROUP BY 
        ThreadID
    ORDER BY 
        COUNT(MessageID) DESC
    LIMIT 1
)
SELECT 
    U.*,
    M.MessageContent,
    M.DateSent
FROM 
    Messages M
JOIN 
    MostMessagesThread T ON M.ThreadID = T.ThreadID
JOIN 
    User U ON U.UserID = M.UserIDSender
ORDER BY 
    M.DateSent ASC;
```

````python
######## Data Preparation for further analysis ############


### STEPS:
#1. Ensure data quality (datatypes, null values etc) 
#2. Derive categorical columns for further analysis and predictive modelling.

from datetime import datetime
import numpy as np

# Clean column names to remove potential issues with spaces or formatting
data.columns = data.columns.str.strip()

# 1. Derive date-related columns
data['time_of_post'] = pd.to_datetime(data['time_of_post'])
data['date'] = data['time_of_post'].dt.date
data['year'] = data['time_of_post'].dt.year
data['month'] = data['time_of_post'].dt.month
data['day'] = data['time_of_post'].dt.day

# Categorize time_of_day
def categorize_time(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    else:
        return "Evening"

data['time_of_day'] = data['time_of_post'].dt.hour.apply(categorize_time)

# 2. Extract geo-info from latitude and longitude (using approximation)
def derive_geo_info(lat, lon):
    if -38 <= lat <= -37 and 144 <= lon <= 145:
        return "Melbourne Area"
    elif -34 <= lat <= -33 and 151 <= lon <= 152:
        return "Sydney Area"
    else:
        return "Other"

data['geo_info'] = data.apply(lambda row: derive_geo_info(row['latitude'], row['longitude']), axis=1)

# 3. Categorize number_of_tradies
def categorize_tradies(count):
    if count < 1000:
        return "Under 1K"
    elif 1000 <= count < 5000:
        return "1-5K"
    elif 5000 <= count < 10000:
        return "5-10K"
    else:
        return "Above 10K"

data['tradies_category'] = data['number_of_tradies'].apply(categorize_tradies)

# 4. Categorize number_of_impressions
def categorize_impressions(count):
    if count < 500:
        return "Low"
    elif 500 <= count < 1000:
        return "Medium"
    else:
        return "High"


# 5. Handle null values in number_of_impressions by imputing median based on geo_info and tradies_category
data['number_of_impressions'] = data.groupby(['geo_info', 'tradies_category'])['number_of_impressions'].transform(
    lambda grp: grp.fillna(grp.median())
)

data['impressions_category'] = data['number_of_impressions'].apply(lambda x: categorize_impressions(x) if not pd.isnull(x) else np.nan)
````

````python
#### Feature importance and predictive modelling


import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Function to calculate VIF
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Function to extract metrics from classification report
def extract_metrics(classification_report):
    lines = classification_report.split("\n")
    metrics_line = lines[7].split()  # Line containing metrics for Class 1
    return {
        "Precision": float(metrics_line[2]),
        "Recall": float(metrics_line[3]),
        "F1-Score": float(metrics_line[4]),
    }

# Function to train and evaluate a model
def train_and_evaluate_model(model, features, X, y, class_weight=None):
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X[features], y, test_size=0.2, random_state=42, stratify=y)
    
    # Scale numerical features
    scaler = StandardScaler()
    numerical_features = ['latitude', 'longitude', 'number_of_tradies', 'number_of_impressions']
    numerical_in_features = [col for col in numerical_features if col in features]
    if numerical_in_features:
        X_train[numerical_in_features] = scaler.fit_transform(X_train[numerical_in_features])
        X_test[numerical_in_features] = scaler.transform(X_test[numerical_in_features])
    
    # Train the model
    model.fit(X_train, y_train, sample_weight=y_train.map(class_weight) if class_weight else None)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    # Evaluation metrics
    classification_rep = classification_report(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    
    # Feature importance
    feature_importances = pd.DataFrame({
        'Feature': features,
        'Importance': model.feature_importances_
    }).sort_values(by='Importance', ascending=False)
    
    return classification_rep, roc_auc, feature_importances

# Derived and numerical features
derived_features = ['geo_info', 'tradies_category', 'impressions_category', 'time_of_day', 'day_name']
numerical_features = ['latitude', 'longitude', 'number_of_tradies', 'number_of_impressions']
combined_features = derived_features + numerical_features

# Encode categorical features
label_encoders = {col: LabelEncoder() for col in derived_features}
for col in derived_features:
    data[col] = label_encoders[col].fit_transform(data[col])

# Prepare X and y
X = data[combined_features].copy()
y = data['accepted']

# Step 1: VIF Analysis and Feature Reduction
vif_data = calculate_vif(X)
high_vif_features = vif_data[vif_data['VIF'] > 10]['Feature'].tolist()
reduced_features = [feature for feature in combined_features if feature not in high_vif_features]

# Step 2: Train Random Forest
rf_model = RandomForestClassifier(class_weight="balanced", random_state=42)
rf_classification_rep, rf_roc_auc, rf_feature_importances = train_and_evaluate_model(rf_model, reduced_features, X, y)

# Step 3: Train Gradient Boosting
gbc_model = GradientBoostingClassifier(random_state=42, n_estimators=200, learning_rate=0.1, max_depth=5)
class_weight_gb = {0: 1, 1: y.value_counts()[0] / y.value_counts()[1]}
gb_classification_rep, gb_roc_auc, gb_feature_importances = train_and_evaluate_model(gbc_model, reduced_features, X, y, class_weight=class_weight_gb)

# Step 4: Compare Results
rf_metrics = extract_metrics(rf_classification_rep)
rf_metrics["ROC-AUC"] = rf_roc_auc

gb_metrics = extract_metrics(gb_classification_rep)
gb_metrics["ROC-AUC"] = gb_roc_auc

comparison = pd.DataFrame({
    "Metric": ["Precision (Class 1)", "Recall (Class 1)", "F1-Score (Class 1)", "ROC-AUC"],
    "Random Forest": [rf_metrics["Precision"], rf_metrics["Recall"], rf_metrics["F1-Score"], rf_metrics["ROC-AUC"]],
    "Gradient Boosting": [gb_metrics["Precision"], gb_metrics["Recall"], gb_metrics["F1-Score"], gb_metrics["ROC-AUC"]]
})

# # Step 5: Display Findings
# import ace_tools as tools
# tools.display_dataframe_to_user(name="Feature Importance - Random Forest", dataframe=rf_feature_importances)
# tools.display_dataframe_to_user(name="Feature Importance - Gradient Boosting", dataframe=gb_feature_importances)
# tools.display_dataframe_to_user(name="Model Comparison - Random Forest vs Gradient Boosting", dataframe=comparison)

# Print Classification Reports for final analysis
print("Random Forest Classification Report:")
print(rf_classification_rep)
print("\nGradient Boosting Classification Report:")
print(gb_classification_rep)
````

# Parameters Influencing the Job Accepted Class (`accepted=1`)

## Feature Importance
After reducing multicollinearity using VIF and analyzing feature importance from both models, the following parameters were identified as the most influential in predicting whether a job would be accepted in this order:

1. **From both Models**:
   - **`number_of_impressions`**: Strongly correlated with job acceptance, indicating job visibility is critical.
   - **`time_of_day`**: Suggests time-sensitive patterns in job posting behavior.
   - **`number_of_tradies`**: Indicates the range of available tradies, impacting job acceptance likelihood..



# Findings and Conclusion

## Findings
### Random Forest Results
- **Precision (Class 1 - Accepted Jobs)**: 0.43
- **Recall (Class 1 - Accepted Jobs)**: 0.33
- **F1-Score (Class 1 - Accepted Jobs)**: 0.37
- **Accuracy**: 0.71
- **ROC-AUC**: 0.59
- **Observation**:
  - Random Forest shows higher overall accuracy but struggles with recall for the minority class (`accepted=1`).

### Gradient Boosting Results
- **Precision (Class 1 - Accepted Jobs)**: 0.43
- **Recall (Class 1 - Accepted Jobs)**: 0.65
- **F1-Score (Class 1 - Accepted Jobs)**: 0.51
- **Accuracy**: 0.68
- **ROC-AUC**: 0.67
- **Observation**:
  - Gradient Boosting significantly improves recall and F1-Score, identifying more true positives (`accepted=1` jobs) while maintaining similar precision.

### Model Strengths
- **Random Forest**:
  - Robust and computationally efficient.
  - Higher accuracy but less effective at handling class imbalance.
- **Gradient Boosting**:
  - Excels at capturing true positives (higher recall).
  - Better overall performance (higher ROC-AUC).

## Conclusion
- **Gradient Boosting** is the better model for this problem, as it:
  - Achieves significantly higher recall and F1-Score for the minority class (`accepted=1`), making it more effective at identifying accepted jobs.
  - Demonstrates superior ROC-AUC, indicating better overall discrimination between classes.

### Recommendation
- Use **Gradient Boosting** for scenarios where identifying accepted jobs (`accepted=1`) is critical.
- Consider **Random Forest** if computational simplicity or overall accuracy is the priority.
